```python
query = """

SELECT
Count(order_id) AS 'Total orders',
(SELECT COUNT(*) FROM orders WHERE orders.order_status = 'delivered') AS 'Delivered Orders',
(SELECT COUNT(*) FROM orders WHERE orders.order_status = 'delivered')*100 / Count(order_id) AS 'Delivered %'
FROM orders

"""
cur.execute(query)
cur.fetchall()

pd.read_sql(query, conn)
```

```python
query = """

SELECT
Count(order_id) AS 'Total orders',
(SELECT COUNT(*) FROM orders WHERE orders.order_status = 'canceled') AS 'Canceled Orders',
(SELECT COUNT(*) FROM orders WHERE orders.order_status = 'canceled')*100 / Count(order_id) AS 'Canceled %'
FROM orders

"""
cur.execute(query)
cur.fetchall()

pd.read_sql(query, conn)
```

This query lists all the product IDs in that order (concatenated into a single string), and returns all orders, total price of orders and the payment type where total price is greater than 100. This may be used to know the total price of each order even it has more than 1 sold product in the invoice and how that orders are paid.

```python
query = '''
SELECT
    COUNT(order_items.order_id) AS total_orders,
    GROUP_CONCAT(order_items.product_id) AS products,
    COUNT(order_items.product_id) AS total_products,
    SUM(order_items.price) AS total_price,
    order_payments.payment_type
FROM order_items
JOIN order_payments ON order_payments.order_id = order_items.order_id
GROUP BY order_items.order_id
HAVING total_price >= 100
'''
pd.read_sql(query,conn)
```

We can use the following query to know about payment process of orders with specific order status.

```python
query = '''
SELECT
    COUNT(order_items.order_id) AS total_orders,
    GROUP_CONCAT(order_items.product_id) AS products,
    COUNT(order_items.product_id) AS total_products,
    SUM(order_items.price) AS total_price,
    order_payments.payment_type,
    orders.order_status
FROM order_items
JOIN order_payments ON order_payments.order_id = order_items.order_id
JOIN orders ON order_items.order_id=orders.order_id
GROUP BY order_items.order_id
HAVING orders.order_status = 'invoiced'
'''
#HAVING orders.order_status = 'approved'
#HAVING orders.order_status = 'shipped'
#HAVING orders.order_status = 'delivered'
#HAVING orders.order_status = 'unavailable'
#HAVING orders.order_status = 'canceled'
#HAVING orders.order_status = 'processing'
#or order_status may be not specified
pd.read_sql(query,conn)
```

```python
query = """SELECT order_id, order_purchase_timestamp,
       order_delivered_customer_date, order_estimated_delivery_date
FROM orders
WHERE order_status = 'delivered';

"""
df = pd.read_sql(query, conn)
df["order_purchase_timestamp"] = pd.to_datetime(df["order_purchase_timestamp"])
df["order_delivered_customer_date"] = pd.to_datetime(df["order_delivered_customer_date"])
df["order_estimated_delivery_date"] = pd.to_datetime(df["order_estimated_delivery_date"])
df["actual_delivery_days"] = (df["order_delivered_customer_date"] - df["order_purchase_timestamp"]).dt.days
df["estimated_delivery_days"] = (df["order_estimated_delivery_date"] - df["order_purchase_timestamp"]).dt.days
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df["order_purchase_timestamp"], df["actual_delivery_days"], label="Actual Delivery Days", marker='o')
plt.plot(df["order_purchase_timestamp"], df["estimated_delivery_days"], label="Estimated Delivery Days", marker='o', linestyle='--')

plt.xlabel("Order Purchase Timestamp")
plt.ylabel("Delivery Days")
plt.title("Comparison of Actual vs. Estimated Delivery")
plt.legend()
plt.grid()
plt.show()
```

Aggregate by Month for Better Visualization

```python
df["purchase_month"] = df["order_purchase_timestamp"].dt.to_period("M")
monthly_summary = df.groupby("purchase_month")[["actual_delivery_days", "estimated_delivery_days"]].mean()

plt.figure(figsize=(12, 6))
plt.plot(monthly_summary.index.astype(str), monthly_summary["actual_delivery_days"], label="Actual Delivery Days", marker='o')
plt.plot(monthly_summary.index.astype(str), monthly_summary["estimated_delivery_days"], label="Estimated Delivery Days", marker='o', linestyle='--')

plt.xlabel("Month")
plt.ylabel("Average Delivery Days")
plt.title("Monthly Comparison of Actual vs. Estimated Delivery")
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```

```python
query = """
SELECT order_id, order_purchase_timestamp, order_approved_at,
       order_delivered_carrier_date, order_delivered_customer_date
FROM orders
WHERE order_status = 'delivered';
"""
df = pd.read_sql(query, conn)
```

```python
df["order_purchase_timestamp"] = pd.to_datetime(df["order_purchase_timestamp"])
df["order_approved_at"] = pd.to_datetime(df["order_approved_at"])
df["order_delivered_carrier_date"] = pd.to_datetime(df["order_delivered_carrier_date"])
df["order_delivered_customer_date"] = pd.to_datetime(df["order_delivered_customer_date"])
```

```python
df["purchase_to_approval"] = (df["order_approved_at"] - df["order_purchase_timestamp"]).dt.total_seconds() / 3600
df["approval_to_carrier"] = (df["order_delivered_carrier_date"] - df["order_approved_at"]).dt.total_seconds() / 3600
df["carrier_to_customer"] = (df["order_delivered_customer_date"] - df["order_delivered_carrier_date"]).dt.total_seconds() / 3600

print(df[["purchase_to_approval", "approval_to_carrier", "carrier_to_customer"]].head())
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df["order_purchase_timestamp"], df["purchase_to_approval"], label="Purchase to Approval", marker='o')
plt.plot(df["order_purchase_timestamp"], df["approval_to_carrier"], label="Approval to Carrier Delivery", marker='o')
plt.plot(df["order_purchase_timestamp"], df["carrier_to_customer"], label="Carrier Delivery to Customer", marker='o')

plt.xlabel("Order Purchase Timestamp")
plt.ylabel("Duration (Hours)")
plt.title("Delivery Timeline Trends")
plt.legend()
plt.grid()
plt.show()
```

Aggregate by Month for Better Visualization

```python
df["purchase_month"] = df["order_purchase_timestamp"].dt.to_period("M")

monthly_summary = df.groupby("purchase_month")[["purchase_to_approval", "approval_to_carrier", "carrier_to_customer"]].mean()

plt.figure(figsize=(12, 6))
plt.plot(monthly_summary.index.astype(str), monthly_summary["purchase_to_approval"], label="Purchase to Approval", marker='o')
plt.plot(monthly_summary.index.astype(str), monthly_summary["approval_to_carrier"], label="Approval to Carrier Delivery", marker='o')
plt.plot(monthly_summary.index.astype(str), monthly_summary["carrier_to_customer"], label="Carrier Delivery to Customer", marker='o')

plt.xlabel("Month")
plt.ylabel("Average Duration (Hours)")
plt.title("Monthly Delivery Timeline Trends")
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
from geopy.geocoders import Nominatim
import pandas as pd

# Initialize geolocator
geolocator = Nominatim(user_agent="customer_geolocator")

# Function to get latitude and longitude of a city
def geocode_city(city, state):
    try:
        location = geolocator.geocode(f"{city}, {state}, Brazil", timeout=10)  # Increased timeout
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except Exception as e:
        print(f"Error geocoding {city}: {e}")
        return None, None

# Apply the function to dataframe
city_customer_counts_df[['latitude', 'longitude']] = city_customer_counts_df.apply(
    lambda row: pd.Series(geocode_city(row['customer_city'], 'SP')), axis=1
)

print(city_customer_counts_df[['customer_city', 'latitude', 'longitude']].head())
```

```python
import plotly.express as px

fig = px.scatter_geo(
    city_customer_counts_df,
    lat="latitude",
    lon="longitude",
    hover_name="customer_city",
    size="customer_count",
    scope="south america",
    title="Customer Distribution in Brazil"
)

fig.show()
```

```python
import folium

brazil_map = folium.Map(location=[-14.2350, -51.9253], zoom_start=4)

# Add customer cities as markers on the map
for _, row in city_customer_counts_df.iterrows():
    if pd.notnull(row['latitude']) and pd.notnull(row['longitude']):
        folium.CircleMarker(
            location=[row['latitude'], row['longitude']],
            radius=8,  # Adjust size
            color="blue",
            fill=True,
            fill_color="blue",
            fill_opacity=0.6,
            popup=f"{row['customer_city']}: {row['customer_count']} customers"
        ).add_to(brazil_map)

brazil_map
```
